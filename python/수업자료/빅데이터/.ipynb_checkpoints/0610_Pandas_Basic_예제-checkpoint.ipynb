```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('X_train.csv', encoding='cp949')
df_test = pd.read_csv('X_test.csv', encoding='cp949')
df = pd.concat([df_train, df_test])
df.head()
```

```python
df['date'] = df['date_time'].str[:10]

df['time'] = pd.to_datetime(df['date_time'])
df['time'] = df['time'].dt.strftime('%H%M')
df['time'] = df['time'].astype('int64')
```

```python
df.head(1)
```

```python
features=[]
```

```python
f = df.groupby('custid')['amount'].agg([('총구매건수', 'count')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['amount'].agg([('평균구매액', 'mean')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['amount'].agg([('총구매액', 'sum')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['discount'].agg([('평균할인금액', 'mean')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['imported'].agg([('평균수입비율', 'mean')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['installment'].agg([('평균할부개월수', 'mean')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
df['total'] = df['amount'] - df['discount']
```

```python
f = df.groupby('custid')['total'].agg([('총실제구매금액', 'sum')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['total'].agg([('평균실제구매금액', 'mean')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
n_brand = df['brand'].nunique()
n_brand
```

```python
def brand(x):
    return x.nunique() / n_brand
f = df.groupby('custid')['brand'].agg([('브랜드다양성', brand)])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
def refund(x):
    if x < 0:
        return '환불'
    else:
        return '구매'
df['refund'] = df['amount'].apply(refund)
```

```python
f = pd.pivot_table(df, index='custid', columns='refund', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
df['datetime'] = pd.to_datetime(df['date_time'])
```

```python
def weekend(x):
    k = x.dayofweek
    if k <= 4 & 1 < x.hour < 19:
        return '평일6시이전'
    else:
        return "그외"
df['weekend'] = df['datetime'].apply(weekend)
```

```python
f = pd.pivot_table(df, index='custid', columns='weekend', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
f = pd.pivot_table(df, index='custid', columns='corner', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = pd.pivot_table(df, index='custid', columns='installment', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = pd.pivot_table(df, index='custid', columns='store', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
df['total_1'] = df['amount'] / df['installment']
```

```python
features.append(f)
```

```python
f = df.groupby('custid')['time'].agg([('평균구매시간', 'mean')])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
import re
male = re.compile('남성|셔츠')
# female = re.compile('여성|란제리|핸드백|부띠끄|부틱|부띠크|미씨|아동복|장신구|유아|침구')
female = re.compile('여성|란제리|핸드백|부띠끄|부틱|부띠크|미씨|아동복|장신구|유아|화장품|침구,수예|침구/수예|로얄부틱|내의란제리|일반식품')

def find_male(x):
    if re.match(male, x):
        return 0
    else:
        return 1


# def find_female(x):
#     if re.match(female, x):
#         return 1
#     else:
#         return 0

    
df['is_male'] = df['pc'].apply(find_male)
# df['is_female'] = df['part'].apply(find_female)
```

```python
f = df.groupby('custid')['is_male'].agg([('남성 단어 수', 'sum')])
f = f.reset_index()
f
```

```python
features.append(f)
```

```python
# 두개의 컬럼의 데이터가 제대로 클러스터링 되는지 확인하는것 
from matplotlib import pyplot as plt
data2 = X_test.loc[:,['총구매건수','평균구매액']]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data2)
labels = kmeans.predict(data2)
plt.scatter(X_test['총구매건수'],X_test['평균구매액'],c = labels)
plt.xlabel('총구매건수')
plt.ylabel('평균구매액')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
import xgboost
from xgboost import XGBClassifier

model = xgboost.XGBRegressor(learning_rate=0.1,
                             max_depth=5,
                             n_estimators=100) 
```

```python
xgb = XGBClassifier(random_state=1, n_job=0) 
score = cross_val_score(xgb, X_train, y_train, cv=5, scoring='roc_auc')

print('score : ', score)
print('mean : ', score.mean())
print('std : ', score.std())
```

```python
pred = xgb.fit(X_train, y_train).predict_proba(X_test)[:, 1]

file_name = 'submission5.csv'

submissions = pd.concat([id_test, pd.Series(pred, name='gender')], axis=1)
submissions.to_csv(file_name, index=False)
print(file_name, 'is ready to submmit!')
```
