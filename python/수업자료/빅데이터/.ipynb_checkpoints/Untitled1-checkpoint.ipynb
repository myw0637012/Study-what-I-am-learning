```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('X_train.csv', encoding='cp949')
df_test = pd.read_csv('X_test.csv', encoding='cp949')
df = pd.concat([df_train, df_test])
df.head()
```

```python
df['date'] = df['date_time'].str[:10]

df['time'] = pd.to_datetime(df['date_time'])
df['time'] = df['time'].dt.strftime('%H%M')
df['time'] = df['time'].astype('int64')
```

```python
n_brand = df['brand'].nunique()
n_brand
```

```python
def brand(x):
    return x.nunique() / n_brand
f = df.groupby('custid')['brand'].agg([('브랜드다양성', brand)])
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
def refund(x):
    if x < 0:
        return '환불'
    else:
        return '구매'
df['refund'] = df['amount'].apply(refund)
```

```python
f = pd.pivot_table(df, index='custid', columns='refund', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
df['datetime'] = pd.to_datetime(df['date_time'])
```

```python
df.info()
```

```python
def weekend(x):
    k = x.dayofweek
    if k <= 4 & 1 < x.hour < 19:
        return '평일6시이전'
    else:
        return "그외"
df['weekend'] = df['datetime'].apply(weekend)
```

```python
f = pd.pivot_table(df, index='custid', columns='weekend', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
f = pd.pivot_table(df, index='custid', columns='corner', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = pd.pivot_table(df, index='custid', columns='installment', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
f = pd.pivot_table(df, index='custid', columns='store', values='amount', aggfunc='count', fill_value=0)
f = f.div(f.sum(axis=1), axis=0).fillna(0)
f = f.reset_index()
f.head()
```

```python
features.append(f)
```

```python
import re
male = re.compile('남성|셔츠|')
# female = re.compile('여성|란제리|핸드백|부띠끄|부틱|부띠크|미씨|아동복|장신구|유아')

def find_male(x):
    if re.match(male, x):
        return 1
    else:
        return 0


# def find_female(x):
#     if re.match(female, x):
#         return 1
#     else:
#         return 0

    
df['is_male'] = df['part'].apply(find_male)
# df['is_female'] = df['part'].apply(find_female)
```

```python
f = df.groupby('custid')['is_male'].agg([('남성 단어 수', 'sum')])
f = f.reset_index()
f
```

```python
features.append(f)
```

```python
X_train = pd.DataFrame({
    'custid' : df_train['custid'].unique()
})

for f in features:
    X_train = pd.merge(X_train, f, how='left')
X_train.head()    
```
