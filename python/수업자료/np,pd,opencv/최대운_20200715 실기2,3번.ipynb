# 최대운_20200715 실기2,3번

```python
#회로도 부품 검출
import random as rng
import cv2
import numpy as np

src_2 = cv2.imread("images/16_NG.bmp", cv2.IMREAD_COLOR)
src_2 = cv2.pyrDown(src_2)
src_copy = src_2.copy()

gray2 = cv2.cvtColor(src_2, cv2.COLOR_BGR2GRAY)
ret2, binary_2 = cv2.threshold(gray2, 80,255,cv2.THRESH_BINARY) #정상적으로 추출 완료
kernel = np.ones((7,7), np.uint8)

#좌측이미지부터 처리============================================================================
img_left = binary_2[206:481, 35:205]#y1:y2, x1:x2
img_left = cv2.morphologyEx(img_left, cv2.MORPH_CLOSE, kernel)
contours_l, hierarchy_l = cv2.findContours(img_left, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

mu = [None]*len(contours_l)
for i in range(len(contours_l)):
    mu[i] = cv2.moments(contours_l[i])#중심점

mc = [None]*len(contours_l)

src_left = src_2.copy()
src_left = src_left[206:481, 35:205]

for i in range(len(contours_l)) :
    mc[i] = (mu[i]['m10'] / (mu[i]['m00'] + 1e-5), mu[i]['m01'] / (mu[i]['m00'] + 1e-5))

left_i = 0
    
for i in range(len(contours_l)) :
    c_area = cv2.contourArea(contours_l[i])
    okcolor = (0, 255, 0)
    ngcolor = (0, 0, 255)

    if c_area > 4000.0 :
        cv2.drawContours(src_left, contours_l, i, okcolor, 2)
        cv2.circle(src_left, (int(mc[i][0]), int(mc[i][1])), 4, okcolor, -1)
        cv2.putText(src_left,"{} : {}".format(i,c_area) ,tuple(contours_l[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,255,0),1)
        
    elif c_area < 1000 and c_area > 800 :
        cv2.drawContours(src_left, contours_l, i, okcolor, 2)
        cv2.circle(src_left, (int(mc[i][0]), int(mc[i][1])), 4, okcolor, -1)
        cv2.putText(src_left,"{} : {}".format(i,c_area) ,tuple(contours_l[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,255,0),1)

    else:#NG일경우
        cv2.drawContours(src_left, contours_l, i, ngcolor, 2)
        cv2.circle(src_left, (int(mc[i][0]), int(mc[i][1])), 4, ngcolor, -1)
        cv2.putText(src_left,"{} : {}".format(i,c_area) ,tuple(contours_l[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,0,255),1)
        left_i += 1


#중앙이미지 처리============================================================================
img_center = binary_2[394:522, 210:1119]#y1:y2, x1:x2
img_center = cv2.morphologyEx(img_center, cv2.MORPH_CLOSE, kernel)
contours_c, hierarchy_c = cv2.findContours(img_center, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

mu = [None]*len(contours_c)
for i in range(len(contours_c)):
    mu[i] = cv2.moments(contours_c[i])#중심점

mc = [None]*len(contours_c)

src_center = src_2.copy()
src_center = src_center[394:522, 210:1119]
for i in range(len(contours_c)) :
    mc[i] = (mu[i]['m10'] / (mu[i]['m00'] + 1e-5), mu[i]['m01'] / (mu[i]['m00'] + 1e-5))

center_i = 0
    
for i in range(len(contours_c)) :
    c_area = cv2.contourArea(contours_c[i])
    okcolor = (0, 255, 0)
    ngcolor = (0, 0, 255)

    cv2.drawContours(src_center, contours_c, i, okcolor, 2)
    cv2.circle(src_center, (int(mc[i][0]), int(mc[i][1])), 4, okcolor, -1)
    cv2.putText(src_center,"{} : {}".format(i,c_area) ,tuple(contours_c[i][0][0]),
           cv2.FONT_HERSHEY_COMPLEX,0.4, (0,255,0),1)
    
    if c_area > 3450 :
        cv2.drawContours(src_center, contours_c, i, okcolor, 2)
        cv2.circle(src_center, (int(mc[i][0]), int(mc[i][1])), 4, okcolor, -1)
        cv2.putText(src_center,"{} : {}".format(i,c_area) ,tuple(contours_c[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,255,0),1)
        
    else:#NG일경우
        cv2.drawContours(src_center, contours_c, i, ngcolor, 2)
        cv2.circle(src_center, (int(mc[i][0]), int(mc[i][1])), 4, ngcolor, -1)
        cv2.putText(src_center,"{} : {}".format(i,c_area) ,tuple(contours_c[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,0,255),1)        
        center_i += 1
        
        
#우측이미지 처리============================================================================
img_right = binary_2[241:464, 1149:1291]#y1:y2, x1:x2
img_right = cv2.morphologyEx(img_right, cv2.MORPH_CLOSE, kernel)
contours_r, hierarchy_r = cv2.findContours(img_right, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

mu = [None]*len(contours_r)
for i in range(len(contours_r)):
    mu[i] = cv2.moments(contours_r[i])#중심점

mc = [None]*len(contours_r)

src_right = src_2.copy()
src_right = src_right[241:464, 1149:1291]

for i in range(len(contours_r)) :
    mc[i] = (mu[i]['m10'] / (mu[i]['m00'] + 1e-5), mu[i]['m01'] / (mu[i]['m00'] + 1e-5))

    
right_i = 0
    
for i in range(len(contours_r)) :
    c_area = cv2.contourArea(contours_r[i])
    okcolor = (0, 255, 0)
    ngcolor = (0, 0, 255)

    if c_area > 3750.0 : 
        cv2.drawContours(src_right, contours_r, i, okcolor, 2)
        cv2.circle(src_right, (int(mc[i][0]), int(mc[i][1])), 4, okcolor, -1)
        cv2.putText(src_right,"{} : {}".format(i,c_area) ,tuple(contours_r[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,255,0),1)
        
    elif c_area < 1000 and c_area > 800 :
        cv2.drawContours(src_right, contours_r, i, okcolor, 2)
        cv2.circle(src_right, (int(mc[i][0]), int(mc[i][1])), 4, okcolor, -1)
        cv2.putText(src_right,"{} : {}".format(i,c_area) ,tuple(contours_r[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,255,0),1)

    else:#NG일경우
        cv2.drawContours(src_right, contours_r, i, ngcolor, 2)
        cv2.circle(src_right, (int(mc[i][0]), int(mc[i][1])), 4, ngcolor, -1)
        cv2.putText(src_right,"{} : {}".format(i,c_area) ,tuple(contours_r[i][0][0]),
               cv2.FONT_HERSHEY_COMPLEX,0.4, (0,0,255),1)
        right_i += 1

        
src_copy[206:481, 35:205] = src_left
src_copy[394:522, 210:1119] = src_center
src_copy[241:464, 1149:1291] = src_right


result = left_i + center_i + right_i

if result == 0:
    cv2.putText(src_copy,"OK",(452,632), cv2.FONT_HERSHEY_COMPLEX,3, (0,255,0),1)
else:
    cv2.putText(src_copy,"NG",(452,632), cv2.FONT_HERSHEY_COMPLEX,3, (0,0,255),1)

cv2.rectangle(src_copy,(28,183) ,(200,491) ,(0,255,0),2)
cv2.rectangle(src_copy,(210,396) ,(1105,500) ,(0,255,0),2)
cv2.rectangle(src_copy,(1135,183) ,(1270,491) ,(0,255,0),2)
    
cv2.imshow("src2",src_copy)
cv2.waitKey(0)


cv2.destroyAllWindows()
```
